```python
wc_name_1 = ['wc1_0711','wc1_0801','wc1_0825','wc1_1001','wc1_1012','wc1_1112','wc1_1127','wc1_0226']
wc_name_2 = ['wc2_0711','wc2_0801','wc2_0825','wc2_1001','wc2_1012','wc2_1112','wc2_1127','wc2_0226']
```
